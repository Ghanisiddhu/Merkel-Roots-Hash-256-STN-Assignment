```python
import bcrypt
import hashlib

my_name = "Usman Ghani"

# Generate bcrypt hash
salt = bcrypt.gensalt()
bcrypt_hash = bcrypt.hashpw(my_name.encode(), salt)
print("Bcrypt Hash: ", bcrypt_hash)

# Generate various other hashes
md5_hash = hashlib.new('md5', my_name.encode()).hexdigest()
print("MD5 Hash: ", md5_hash)

sha1_hash = hashlib.new('sha1', my_name.encode()).hexdigest()
print("SHA1 Hash: ", sha1_hash)

sha256_hash = hashlib.new('sha256', my_name.encode()).hexdigest()
print("SHA256 Hash: ", sha256_hash)

sha512_hash = hashlib.new('sha512', my_name.encode()).hexdigest()
print("SHA512 Hash: ", sha512_hash)

sha3_hash = hashlib.new('sha3_224', my_name.encode()).hexdigest()
print("SHA3 Hash: ", sha3_hash)

blake2_hash = hashlib.new('blake2b', my_name.encode()).hexdigest()
print("BLAKE2 Hash: ", blake2_hash)
```

```python
from Crypto.Hash import RIPEMD160

my_name = "Usman Ghani"

# RIPEMD-160 hash object
ripemd_hash = RIPEMD160.new()

# hash object data to be hashed
ripemd_hash.update(my_name.encode())

# hash value
RIPEMD_160 = ripemd_hash.hexdigest()

print("RIPEMD-160 Hash: ", RIPEMD_160)
```

```python
import hashlib
import requests

def calculate_sha256(file_content):
    """
    Calculate SHA-256 hash of file content.
    """
    sha256_hash = hashlib.sha256()
    sha256_hash.update(file_content)
    return sha256_hash.hexdigest()

def download_file(file_url):
    """
    Download file content from the provided URL.
    """
    response = requests.get(file_url)
    if response.status_code == 200:
        return response.content
    else:
        print("Failed to download the file.")
        return None

def main():
    file_url1 = 'https://drive.google.com/uc?id=1Fi9CUIo4UTvGFXQUwOrMnQxVW5Jf95CM'
    file_url2 = 'https://drive.google.com/uc?id=15kvWgwZNHWx7FxJeLrcU7KHDN8sdNXe8'

    # Download file 1 content
    file1_content = download_file(file_url1)
    # Download file 2 content
    file2_content = download_file(file_url2)

    if file1_content and file2_content:
        # Calculate SHA-256 hashes
        file1_hash = calculate_sha256(file1_content)
        file2_hash = calculate_sha256(file2_content)

        print("SHA-256 Hash of file 1:", file1_hash)
        print("SHA-256 Hash of file 2:", file2_hash)

        # Compare hashes
        differences = sum(bin(int(h1, 16) ^ int(h2, 16)).count('1') for h1, h2 in zip(file1_hash, file2_hash))
        total_bits = len(file1_hash) * 4  # Each hex character represents 4 bits
        percentage_difference = (differences / total_bits) * 100

        print("Percentage of bits different (Avalanche Effect):", percentage_difference)
    else:
        print("One or both files failed to download.")

if __name__ == "__main__":
    main()
```

```python
import requests
import hashlib

def download_file(url, filename):
    response = requests.get(url)
    with open(filename, 'wb') as f:
        f.write(response.content)

def calculate_hashes(filename):
    md5_hash = hashlib.md5()
    sha1_hash = hashlib.sha1()

    with open(filename, 'rb') as f:
        while chunk := f.read(4096):
            md5_hash.update(chunk)
            sha1_hash.update(chunk)

    return md5_hash.hexdigest(), sha1_hash.hexdigest()

if __name__ == "__main__":
    # URLs of the files
    file_urls = {
        'message1': 'https://drive.google.com/uc?export=download&id=13f7cFusdVTCFlwtBO0gxHRUAP_GIPTej',
        'message2': 'https://drive.google.com/uc?export=download&id=1YdhdMYa2BXMwMT7KnPk1xSp0nWswxvJD'
    }

    # Download and calculate hashes
    for file_name, url in file_urls.items():
        print(f"Downloading {file_name}.bin...")
        download_file(url, f"{file_name}.bin")
        print(f"Calculating hashes for {file_name}.bin...")
        md5_hash, sha1_hash = calculate_hashes(f"{file_name}.bin")
        print(f"MD5 hash of {file_name}.bin: {md5_hash}")
        print(f"SHA-1 hash of {file_name}.bin: {sha1_hash}")
        print()
```

```python
import hashlib
import math

def calculate_block_hash(string):
    return hashlib.sha256(string.encode()).hexdigest()

def calculate_merkle_root(strings):
    # Calculate block hashes
    block_hashes = [calculate_block_hash(string) for string in strings]

    # Create list to hold current level of tree
    current_level = block_hashes

    # Continue until we get the root
    while len(current_level) > 1:
        next_level = []
        for i in range(0, len(current_level), 2):
            # Concatenate two hashes
            concatenated_hash = current_level[i] + current_level[i+1] if i+1 < len(current_level) else current_level[i]
            # Calculate hash of concatenated hashes
            block_hash = hashlib.sha256(concatenated_hash.encode()).hexdigest()
            next_level.append(block_hash)
        current_level = next_level

    return current_level[0] if current_level else None

# Generate eight random strings
import random
import string

random_strings = [''.join(random.choices(string.ascii_letters, k=10)) for _ in range(8)]

# Print generated strings
print("Generated Strings:")
for string in random_strings:
    print(string)

# Calculate Merkle root
merkle_root = calculate_merkle_root(random_strings)

# Print Merkle root
print("\nMerkle Root:")
print(merkle_root)
```

```python
import hashlib
import requests

def download_file_from_google_drive(id, destination):
    URL = "https://drive.google.com/uc?export=download"

    session = requests.Session()

    response = session.get(URL, params={'id': id}, stream=True)
    token = None
    for key, value in response.cookies.items():
        if key.startswith('download_warning'):
            token = value
            break

    if token:
        params = {'id': id, 'confirm': token}
        response = session.get(URL, params=params, stream=True)

    with open(destination, "wb") as f:
        for chunk in response.iter_content(32768):
            if chunk:
                f.write(chunk)

def calculate_block_hash(data):
    return hashlib.sha256(data.encode()).hexdigest()

def calculate_merkle_root(blocks):
    block_hashes = [calculate_block_hash(block) for block in blocks]
    print("Block Hashes:")
    print(block_hashes)

    current_level = block_hashes

    while len(current_level) > 1:
        print("\nLevel Hashes:")
        print(current_level)
        next_level = []
        for i in range(0, len(current_level), 2):
            concatenated_hash = current_level[i] + current_level[i+1] if i+1 < len(current_level) else current_level[i]
            block_hash = hashlib.sha256(concatenated_hash.encode()).hexdigest()
            next_level.append(block_hash)
        current_level = next_level

    print("\nFinal Merkle Root:")
    return current_level[0] if current_level else None

# Download file from Google Drive
file_id = "1ExX6aTwTOsxX7rFJBwaK4egWzS1H-9UF"
destination_file = "data.txt"
download_file_from_google_drive(file_id, destination_file)

# Parse file into eight blocks
with open(destination_file, 'r') as file:
    lines = file.readlines()
    num_blocks = min(len(lines), 8)
    blocks = [lines[i].strip() for i in range(num_blocks)]

# Calculate Merkle root
merkle_root = calculate_merkle_root(blocks)

# Print Merkle root
print(merkle_root)
```
